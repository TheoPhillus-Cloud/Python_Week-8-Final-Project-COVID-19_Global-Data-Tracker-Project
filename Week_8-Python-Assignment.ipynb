```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset 
from sklearn.datasets import load_iris
iris_data = load_iris()

# Convert to a DataFrame
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# Display the first few rows
print("First few rows of the dataset:")
print(df.head())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Display dataset info
print("\nDataset Info:")
print(df.info())

# Clean the dataset 
df.dropna(inplace=True)  # Or df.fillna(value, inplace=True)
```

```python
# Compute basic statistics
print("\nBasic Statistics:")
print(df.describe())

# Group by species and compute mean values
grouped_df = df.groupby("species").mean()
print("\nMean values grouped by species:")
print(grouped_df)

# Identify interesting patterns
print("\nObservations:")
print("Sepal length seems to increase gradually across species.")
print("Petal width shows a significant difference between species.")
```

```python
# Style for better visuals
sns.set_style("whitegrid")

# 1️ Line Chart (for trends)
plt.figure(figsize=(8,5))
df.groupby("species").mean().plot(kind="line", marker="o")
plt.title("Mean Feature Values per Species")
plt.ylabel("Measurement (cm)")
plt.xlabel("Species")
plt.legend(loc="upper left")
plt.show()

# 2️ Bar Chart (comparison across categories)
plt.figure(figsize=(8,5))
sns.barplot(x="species", y="sepal length (cm)", data=df)
plt.title("Average Sepal Length per Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()

# 3️ Histogram (distribution visualization)
plt.figure(figsize=(8,5))
sns.histplot(df["petal width (cm)"], bins=20, kde=True)
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.show()

# 4️ Scatter Plot (relationship between numerical values)
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df)
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()
```
